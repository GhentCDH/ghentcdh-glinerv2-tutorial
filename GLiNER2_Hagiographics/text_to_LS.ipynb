## text to LS

this code will convert the Hagiographics text to LS format. It will first split the text in to different sections, based on the '[number]' pattern. Then it will put the split text in the right LS format

```python
import re
import json
from pathlib import Path

input_file = Path('./test_texts/BHL7120-7125 - clean.txt')
text = input_file.read_text(encoding='utf-8')


segments = re.split(r'(?=\[\d+\])', text)

task = []
for seg in segments:
    seg = seg.strip()
    if seg:
        task.append({"data": {"text": seg}})

output_file = input_file.with_name(input_file.stem + ' - LS.json')

with open (output_file, 'w', encoding='utf-8') as f:
    json.dump(task, f, ensure_ascii=False, indent=2)

print(f'Converted {len(segments)} segments to LS format and saved to {output_file}')
```

## text to Batch example text
This code will convert the Hagiographics texts to a batch example text to run the GLiNER2 adaptor on. It will first split the text in to different sections, based on the '[number]' pattern. Then it will put the split texts into a example text folder.

```python
import re
from pathlib import Path

input_file = Path('./test_texts/BHL2605 - clean.txt')
text = input_file.read_text(encoding='utf-8')

segments = re.split(r'(?=\[\d+\])', text)

# Create output directory
output_dir = Path('./Hagiographics_example_texts')
output_dir.mkdir(exist_ok=True)

# Save each segment as a separate .txt file
count = 0
for i, seg in enumerate(segments, start=1):
    seg = seg.strip()
    if seg:
        output_file = output_dir / f'{input_file.stem}_{i}.txt'
        with open(output_file, 'w', encoding='utf-8') as f:
            f.write(seg)
        count += 1

print(f'Converted {count} segments to individual text files and saved to {output_dir}')
```
